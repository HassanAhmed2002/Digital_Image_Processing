```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color
from skimage.transform import rescale, resize
```

```python
#TASK 1

image = color.rgb2gray(data.astronaut())
image_rescaled = rescale(image, 0.25)

fig, axes = plt.subplots(nrows=1, ncols=2)

ax = axes.ravel()

ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original")

ax[1].imshow(image_rescaled, cmap='gray')
ax[1].set_title("Rescaled to 1/4 of original")

ax[0].set_xlim(0, 512)
ax[0].set_ylim(512, 0)
plt.tight_layout()
plt.show()
```

```python
#TASK 2

image = color.rgb2gray(data.astronaut())
image_resized = resize(image, (100, 200))

fig, axes = plt.subplots(nrows=1, ncols=2)

ax = axes.ravel()

ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original")

ax[1].imshow(image_resized, cmap='gray')
ax[1].set_title("Resized to 100x200")

ax[0].set_xlim(0, 512)
ax[0].set_ylim(512, 0)
plt.tight_layout()
plt.show()
```

```python
#TASK 3
from scipy import interpolate

x = np.linspace(0, 10, num=10, endpoint=True)
y = [5,3,7,10,5,7,10,10,7,8]
f_linear = interpolate.interp1d(x, y, 'linear')
f_quadratic = interpolate.interp1d(x, y, 'quadratic')
f_cubic = interpolate.interp1d(x, y, 'cubic')
xnew = np.linspace(0, 10, num=100, endpoint=True)
ynew_linear = f_linear(xnew)
ynew_quadratic = f_quadratic(xnew)
ynew_cubic = f_cubic(xnew)
```

```python
plt.scatter(x,y, color='blue', label='Original')
plt.plot(xnew, ynew_linear, color='yellow', label='Linear 1D')
plt.plot(xnew, ynew_quadratic, color='green', label='Quadratic 1D')
plt.plot(xnew, ynew_cubic, color='red', label='Cubic 1D')
plt.legend()
plt.show()
```

```python
#TASK 4

x = np.arange(-5.01, 5.01, 0.25)
y = np.arange(-5.01, 5.01, 0.25)
x_img, y_img = np.meshgrid(x, y)
z = np.sin(x_img**2+y_img**2)

f_linear_2d = interpolate.interp2d(x_img, y_img, z, kind='linear')
f_cubic_2d = interpolate.interp2d(x_img, y_img, z, kind='cubic')
f_quintic_2d = interpolate.interp2d(x_img, y_img, z, kind='quintic')
```

```python
xnew = np.arange(-5.01, 5.01, 1e-2)
ynew = np.arange(-5.01, 5.01, 1e-2)
z_linear_2d = f_linear_2d(xnew, ynew)
z_cubic_2d = f_cubic_2d(xnew, ynew)
z_quintic_2d = f_quintic_2d(xnew, ynew)
```

```python
plt.scatter(x, z[0, :], color='blue', label='Original')
plt.plot(xnew, z_linear_2d[0, :], color='yellow', label='Linear 2D')
plt.plot(xnew, z_cubic_2d[0, :], color='green', label='Cubic 2D')
plt.plot(xnew, z_quintic_2d[0, :], color='red', label='Quintic 2D')
plt.legend()
plt.show()
```
